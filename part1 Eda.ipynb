# Filling missing values

```python
df['loan_int_rate'].fillna(df['loan_int_rate'].median(),inplace=True)
```

```python
df['person_emp_length'].fillna(df['person_emp_length'].median(),inplace=True)
```

```python
df=df[df['person_age']<80]
df=df[df['person_emp_length']<50]
```

# Univariate Analysis

# Numerical Columns

```python
top10 = df.sort_values(by="person_income", ascending=False).head(10)
plt.figure()
plt.bar(top10.index.astype(str), top10["person_income"])
plt.title("Top 10 Person Income")
plt.xlabel("Person ID")
plt.ylabel("Person Income")
plt.xticks(rotation=45)
plt.show()
```

```python
top10 = df.sort_values(by="person_emp_length", ascending=False).head(10)
plt.figure()
plt.bar(top10.index.astype(str), top10["person_emp_length"])
plt.title("Top 10 Person Employment Length")
plt.xlabel("Person ID")
plt.ylabel("Person Employment Length")
plt.xticks(rotation=45)
plt.show()
```

# Categorical Columns

```python
cat_cols = [
    'person_home_ownership','loan_intent',
    'loan_grade','cb_person_default_on_file'
]

for col in cat_cols:
    order = df[col].value_counts().index
    plt.figure()
    sns.countplot(x=col, data=df, order=order)
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

# Bivariate Analysis

```python
grade_default = df.groupby('loan_grade')['loan_status'].mean().sort_values()

grade_default.plot(kind='bar')
plt.title("Default Rate by Loan Grade")
plt.ylabel("Default Rate")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='loan_percent_income', data=df)
plt.title("Loan Percent Income vs Loan Status")
plt.show()
```

```python
sns.barplot(
    x='cb_person_default_on_file',
    y='loan_status',
    data=df
)
plt.title("Default Rate by Previous Default History")
plt.ylabel("Default Rate")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='loan_int_rate', data=df)
plt.title("Interest Rate vs Loan Status")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='person_emp_length', data=df)
plt.title("Employment Length vs Loan Status")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='person_income', data=df)
plt.yscale('log')
plt.title("Income vs Loan Status (Log Scale)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['loan_amnt'], bins=30, kde=True)
plt.title("Distribution of Loan Amount")
plt.show()

plt.figure(figsize=(8,3))
sns.boxplot(x=df['loan_amnt'])
plt.title("Loan Amount Outliers")
plt.show()
```

```python
sns.countplot(x='person_home_ownership', hue='loan_status', data=df)
plt.title("Home Ownership vs Loan Status")
plt.xticks(rotation=30)
plt.show()
```

```python
sns.countplot(x='loan_intent', hue='loan_status', data=df)
plt.title("Loan Intent vs Loan Status")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='loan_grade', hue='loan_status', data=df)
plt.title("Loan Grade vs Loan Status")
plt.show()
```

```python
sns.countplot(x='cb_person_default_on_file', hue='loan_status', data=df)
plt.title("Credit Default History vs Loan Status")
plt.show()
```
