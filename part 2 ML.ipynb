```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('clean_data.csv')
```

```python
print("df.shape:", df.shape)
print("columns:", df.columns.tolist())

df.head()
```

# Split in x and y

```python
x = df[['person_age','person_income','person_emp_length','loan_amnt','loan_int_rate','loan_percent_income','cb_person_cred_hist_length']]
y = df['loan_status']
```

```python
#train and test split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
col=['person_age','person_income','person_emp_length','loan_amnt','loan_int_rate','loan_percent_income','cb_person_cred_hist_length']
```

```python
#standardizing the X Train and Test
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train_transformed = pd.DataFrame(scale.fit_transform(x_train),columns = col, index = x_train.index)
x_test_transformed = pd.DataFrame(scale.fit_transform(x_test), columns = col,index = x_test.index)
print(x_train_transformed)
```

```python
x_train_transformed.info()
```

# Model Building

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
LR_classifier = LogisticRegression()
LR_classifier.fit(x_train_transformed,y_train)
```

```python
y_test_pred = LR_classifier.predict(x_test_transformed)
log_accuracy = (accuracy_score(y_test,y_test_pred))
print(classification_report(y_test,y_test_pred))
log_accuracy
```

```python
#using KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
KNN_classifier = KNeighborsClassifier()
KNN_classifier.fit(x_train_transformed,y_train)
```

```python
y_test_pred = KNN_classifier.predict(x_test_transformed)
knn_accuracy=(accuracy_score(y_test,y_test_pred))
print(classification_report(y_test,y_test_pred))
knn_accuracy
```

```python
# using Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
DT_classifier = DecisionTreeClassifier()
DT_classifier.fit(x_train_transformed,y_train)
```

```python
y_test_pred = DT_classifier.predict(x_test_transformed)
DT_accuracy=(accuracy_score(y_test,y_test_pred))
print(classification_report(y_test,y_test_pred))
DT_accuracy
```

```python
# using ensemble approach -- Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
RF_classifier = RandomForestClassifier()
RF_classifier.fit(x_train_transformed,y_train)
```

```python
y_test_pred = RF_classifier.predict(x_test_transformed)
RF_accuracy=(accuracy_score(y_test,y_test_pred))
print(classification_report(y_test,y_test_pred))
RF_accuracy
```

```python
#using SVM
from sklearn.svm import SVC
SV_classifier = SVC()
SV_classifier.fit(x_train_transformed,y_train)
```

```python
y_test_pred = SV_classifier.predict(x_test_transformed)
SV_accuracy=(accuracy_score(y_test,y_test_pred))
print(classification_report(y_test,y_test_pred))
SV_accuracy
```

```python
evaluation = pd.DataFrame({'Algorithms':['Logistic Regression','Decision Tree','Random Forest','SVC','KNN'],"Accuracy":[log_accuracy,DT_accuracy,RF_accuracy,SV_accuracy,knn_accuracy]})
```
